#Analysis

    1. Cities in the northern hemisphere are warmer this time of year
    2. For the random cities chosen, most have medium to high humidity
    3. Wind speeds are somewhat low as of the data snapshot

```python
#Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import random
import json
import requests

from citipy import citipy

from config import weather_key
```

```python
coords = []

for x in range(2000):
    longitude = round(random.uniform(-180,180),4)
    latitude = round(random.uniform(-90,90),4)
    coords.append({"Longitude":longitude, "Latitude":latitude})
    
```

```python
cities = []
unique_cities = []
counter = 1

for coord in coords:
    city = citipy.nearest_city(coord['Latitude'],coord['Longitude'])
    
    if city.city_name not in unique_cities:
        cities.append({"Counter":counter, "CityName":city.city_name, "CountryCode":city.country_code, \
                       "Latitude":coord['Latitude'], "Longitude":coord['Longitude'], "Temperature":0, \
                      "Humidity":0, "Clouds":0, "WindSpeed":0})
        counter += 1
        unique_cities.append(city.city_name)
        
    if len(unique_cities) == 500:
        break
```

```python
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&"
```

```python
for city in cities:
    try:
        query_url = url + "appid=" + weather_key + "&q=" + city['CityName']
        print(query_url)
        weather_response = requests.get(query_url)
        weather_json = weather_response.json()
        city["Temperature"] = weather_json['main']['temp_max']
        city["Humidity"] = weather_json['main']['humidity']
        city["Clouds"] = weather_json['clouds']['all']
        city["WindSpeed"] = weather_json['wind']['speed']
    except:
        print(f"City not found: {city['CityName']}({city['Counter']})")
```

```python
cities_df = pd.DataFrame(cities)
```

```python
cities_df.to_csv("WeatherPy_Data.csv")
```

```python
x_axis = cities_df['Latitude']
data = cities_df['Temperature']

plt.title("City Latitude vs. Max Temperature (06/20/18)")
plt.xlabel("Latitude")
plt.ylabel("Max Temperature (F)")

plt.scatter(x_axis, data, marker="o", facecolors="blue", edgecolors="black", alpha=0.75)
plt.ylim(0, 120)
plt.xlim(-90,90)
plt.grid()
plt.show()
```
